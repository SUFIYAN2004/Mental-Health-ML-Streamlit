```python
import kagglehub
path = kagglehub.dataset_download("algozee/teenager-menthal-healy")
```

```python
import pandas as pd
import numpy as np
import os

files_list = os.listdir(path)
actual_filename = files_list[0]
full_file_path = f"{path}/{actual_filename}"
df = pd.read_csv(full_file_path)
df.head()
```

```python
df['platform_usage'].value_counts()
```

```python
df['gender'].value_counts()
```

```python
df['social_interaction_level'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder


a = df['gender']
b = df['platform_usage']
c = df['social_interaction_level']

gender_le = LabelEncoder()
platform_le = LabelEncoder()
Social_le = LabelEncoder()

df['gender'] = gender_le.fit_transform(a)
df['platform_usage'] = platform_le.fit_transform(b)
df['social_interaction_level'] = Social_le.fit_transform(c)
```

```python
df
```

```python
df['depression_label'].value_counts()
```

```python
X = df.iloc[:, :-1].values
y =df.iloc[:, -1].values
```

```python
X[0]
```

```python
y[0]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    max_iter=500,
    class_weight='balanced'
)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

score = classification_report(y_test, y_pred)
print("Classification reports: \n",score)
```

```python
from sklearn.model_selection import cross_val_score

cross = cross_val_score(model, X, y, cv=3)
print("cross: ",cross)
```

```python
print("Mean: ", cross.mean())
print(f"Standard: {cross.std()}")
```
